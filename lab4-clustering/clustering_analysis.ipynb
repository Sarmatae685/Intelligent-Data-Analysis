```python
#імпорти
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import MiniBatchKMeans
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score, adjusted_rand_score

# Підключення до Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Завантаження даних
dataset_path = 'lab4_clustering.csv'
df = pd.read_csv(dataset_path)
```

```python
# Візуалізація початкових даних
plt.figure(figsize=(10, 6))
sns.scatterplot(x='X1', y='X2', hue='not_y', data=df, palette='viridis', s=60)
plt.title("Initial Data with True Labels")
plt.show()
```

```python
# Вибір вхідних ознак
X = df[['X1', 'X2']]
```

```python
# Алгоритм Mini-Batch K-Means
kmeans = MiniBatchKMeans(n_clusters=7, random_state=42)
kmeans_labels = kmeans.fit_predict(X)

# Оцінка Mini-Batch K-Means
kmeans_silhouette = silhouette_score(X, kmeans_labels)
kmeans_ari = adjusted_rand_score(df['not_y'], kmeans_labels)
print(f"Mini-Batch K-Means - Silhouette Score: {kmeans_silhouette:.4f}, ARI: {kmeans_ari:.4f}")

# Візуалізація кластерів Mini-Batch K-Means
plt.figure(figsize=(10, 6))
sns.scatterplot(x='X1', y='X2', hue=kmeans_labels, data=df, palette='viridis', s=60)
plt.title("Mini-Batch K-Means Clustering")
plt.show()
```

```python
# Алгоритм Gaussian Mixture Model
gmm = GaussianMixture(n_components=7, random_state=42)
gmm_labels = gmm.fit_predict(X)

# Оцінка Gaussian Mixture Model
gmm_silhouette = silhouette_score(X, gmm_labels)
gmm_ari = adjusted_rand_score(df['not_y'], gmm_labels)
print(f"Gaussian Mixture Model - Silhouette Score: {gmm_silhouette:.4f}, ARI: {gmm_ari:.4f}")

# Візуалізація кластерів Gaussian Mixture Model
plt.figure(figsize=(10, 6))
sns.scatterplot(x='X1', y='X2', hue=gmm_labels, data=df, palette='viridis', s=60)
plt.title("Gaussian Mixture Model Clustering")
plt.show()
```

```python
# Порівняння результатів
print("\nПорівняння результатів кластеризації:")
print(f"Mini-Batch K-Means - Silhouette Score: {kmeans_silhouette:.4f}, ARI: {kmeans_ari:.4f}")
print(f"Gaussian Mixture Model - Silhouette Score: {gmm_silhouette:.4f}, ARI: {gmm_ari:.4f}")
```
